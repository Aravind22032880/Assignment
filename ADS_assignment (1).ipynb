```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

# Load the Uber stock data
uber_stock_data = pd.read_csv("UBER.csv")

# Convert the 'Date' column to datetime format
uber_stock_data['Date'] = pd.to_datetime(uber_stock_data['Date'])

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12,8))

# Plot the Closing price of Uber stock over time
ax.plot(uber_stock_data['Date'], uber_stock_data['Close'], label='Closing price')

# Plot the High price of Uber stock over time
ax.plot(uber_stock_data['Date'], uber_stock_data['High'], label='High price')

# Plot the Low price of Uber stock over time
ax.plot(uber_stock_data['Date'], uber_stock_data['Low'], label='Low price')

# Set the x-axis label
ax.set_xlabel('Date')

# Set the y-axis label
ax.set_ylabel('Price (USD)')

# Set the title
ax.set_title('Uber Stock Prices')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
df= uber_stock_data
# Create a bar plot of the average closing price by year
df['Year'] = pd.DatetimeIndex(df['Date']).year
df.groupby('Year')['Close'].mean().plot(kind='bar')
plt.title('Average Closing Price by Year')
plt.xlabel('Year')
plt.ylabel('Closing Price')
plt.show()
```

```python
# Create a scatter plot of the volume traded vs. the high price
plt.scatter(df['Volume'], df['High'])
plt.title('Volume Traded vs. High Price')
plt.xlabel('Volume')
plt.ylabel('High Price')
plt.show()
```
